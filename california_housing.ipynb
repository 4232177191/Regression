```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate approximate latitude & longitude (since California_df doesn't have them)
np.random.seed(42)
California_df['latitude'] = np.random.uniform(32.5343, 42.0095, size=len(California_df))   # Random latitudes around California


California_df['longitude'] = np.random.uniform(-124.4096, -114.8183, size=len(California_df))     # Random longitudes around California


# Define KFold cross-validation
num_splits = 5
num_models = 4
kfold = KFold(n_splits=num_splits, shuffle=True)

# Create an array to store RMSE values
rmses = np.zeros((num_models, num_splits))

# Define RMSE function
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Loop through KFold splits
for i, (train_index, test_index) in enumerate(kfold.split(California_df)):

    # Split dataset into training and testing
    train_set = California_df.iloc[train_index]
    test_set = California_df.iloc[test_index]

    # Target variable
    y_test = test_set["MedHouseVal"]

    ## 1️⃣ Baseline Model (Mean Prediction)
    baseline_pred = np.full(len(test_set), train_set["MedHouseVal"].mean())
    rmses[0, i] = root_mean_squared_error(y_test, baseline_pred)

    ## 2️⃣ Simple Linear Regression (SLR) - Using 'AveOccup'
    slr = LinearRegression()
    slr.fit(train_set[['AveOccup']], train_set["MedHouseVal"])
    rmses[1, i] = root_mean_squared_error(y_test, slr.predict(test_set[['AveOccup']]))

    ## 3️⃣ Multiple Linear Regression (MLR) - Using 'AveOccup', 'AveBedrms'
    mlr = LinearRegression()
    mlr.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[2, i] = root_mean_squared_error(y_test, mlr.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 4️⃣ K-Nearest Neighbors (KNN) Regression - Using 'latitude' & 'longitude'
    knn = KNeighborsRegressor(n_neighbors=10)
    knn.fit(train_set[['latitude', 'longitude']], train_set["MedHouseVal"])
    rmses[3, i] = root_mean_squared_error(y_test, knn.predict(test_set[['latitude', 'longitude']]))

# Display the RMSE results
print("RMSE values for each model across folds:\n", rmses)
print("\nMean RMSE for each model:\n", np.mean(rmses, axis=1))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate approximate latitude & longitude (since California_df doesn't have them)
np.random.seed(42)
California_df['latitude'] = np.random.uniform(32.5343, 42.0095, size=len(California_df))   # Random latitudes around California
California_df['longitude'] = np.random.uniform(-124.4096, -114.8183, size=len(California_df))     # Random longitudes around California

# Define KFold cross-validation
num_splits = 5
num_models = 5  # Now we have 5 models
kfold = KFold(n_splits=num_splits, shuffle=True)

# Create an array to store RMSE values
rmses = np.zeros((num_models, num_splits))

# Define RMSE function
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Loop through KFold splits
for i, (train_index, test_index) in enumerate(kfold.split(California_df)):

    # Split dataset into training and testing
    train_set = California_df.iloc[train_index]
    test_set = California_df.iloc[test_index]

    # Target variable
    y_test = test_set["MedHouseVal"]

    ## 1️⃣ Baseline Model (Mean Prediction)
    baseline_pred = np.full(len(test_set), train_set["MedHouseVal"].mean())
    rmses[0, i] = root_mean_squared_error(y_test, baseline_pred)

    ## 2️⃣ Simple Linear Regression (SLR) - Using 'AveOccup'
    slr = LinearRegression()
    slr.fit(train_set[['AveOccup']], train_set["MedHouseVal"])
    rmses[1, i] = root_mean_squared_error(y_test, slr.predict(test_set[['AveOccup']]))

    ## 3️⃣ Multiple Linear Regression (MLR) - Using 'AveOccup', 'AveBedrms'
    mlr = LinearRegression()
    mlr.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[2, i] = root_mean_squared_error(y_test, mlr.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 4️⃣ Huber Regression - Using 'AveOccup' and 'AveBedrms' (instead of Linear Regression)
    huber = HuberRegressor()
    huber.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[3, i] = root_mean_squared_error(y_test, huber.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 5️⃣ K-Nearest Neighbors (KNN) Regression - Using 'latitude' & 'longitude' with Manhattan Distance
    knn = KNeighborsRegressor(n_neighbors=10, metric='manhattan')  # Use 'manhattan' distance
    knn.fit(train_set[['latitude', 'longitude']], train_set["MedHouseVal"])
    rmses[4, i] = root_mean_squared_error(y_test, knn.predict(test_set[['latitude', 'longitude']]))

# Display the RMSE results
print("RMSE values for each model across folds:\n", rmses)
print("\nMean RMSE for each model:\n", np.mean(rmses, axis=1))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, HuberRegressor, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate approximate latitude & longitude (since California_df doesn't have them)
np.random.seed(42)
California_df['latitude'] = np.random.uniform(32.5343, 42.0095, size=len(California_df))   # Random latitudes around California
California_df['longitude'] = np.random.uniform(-124.4096, -114.8183, size=len(California_df))     # Random longitudes around California

# Define KFold cross-validation
num_splits = 5
num_models = 8  # Now we have 8 models
kfold = KFold(n_splits=num_splits, shuffle=True)

# Create an array to store RMSE values
rmses = np.zeros((num_models, num_splits))

# Define RMSE function
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Loop through KFold splits
for i, (train_index, test_index) in enumerate(kfold.split(California_df)):

    # Split dataset into training and testing
    train_set = California_df.iloc[train_index]
    test_set = California_df.iloc[test_index]

    # Target variable
    y_test = test_set["MedHouseVal"]

    ## 1️⃣ Baseline Model (Mean Prediction)
    baseline_pred = np.full(len(test_set), train_set["MedHouseVal"].mean())
    rmses[0, i] = root_mean_squared_error(y_test, baseline_pred)

    ## 2️⃣ Simple Linear Regression (SLR) - Using 'AveOccup'
    slr = LinearRegression()
    slr.fit(train_set[['AveOccup']], train_set["MedHouseVal"])
    rmses[1, i] = root_mean_squared_error(y_test, slr.predict(test_set[['AveOccup']]))

    ## 3️⃣ Multiple Linear Regression (MLR) - Using 'AveOccup', 'AveBedrms'
    mlr = LinearRegression()
    mlr.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[2, i] = root_mean_squared_error(y_test, mlr.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 4️⃣ Huber Regression - Using 'AveOccup' and 'AveBedrms'
    huber = HuberRegressor()
    huber.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[3, i] = root_mean_squared_error(y_test, huber.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 5️⃣ Ridge Regression - Using 'AveOccup' and 'AveBedrms'
    ridge = Ridge(alpha=1.0)  # You can adjust alpha for regularization strength
    ridge.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[4, i] = root_mean_squared_error(y_test, ridge.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 6️⃣ Lasso Regression - Using 'AveOccup' and 'AveBedrms'
    lasso = Lasso(alpha=0.1)  # You can adjust alpha for regularization strength
    lasso.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[5, i] = root_mean_squared_error(y_test, lasso.predict(test_set[['AveOccup', 'AveBedrms']]))

    ## 7️⃣ ElasticNet Regression - Using 'AveOccup' and 'AveBedrms'
    elastic_net = ElasticNet(alpha=1.0, l1_ratio=0.5)  # You can adjust alpha and l1_ratio
    elastic_net.fit(train_set[['AveOccup', 'AveBedrms']], train_set["MedHouseVal"])
    rmses[6, i] = root_mean_squared_error(y_test, elastic_net.predict(test_set[['AveOccup', 'AveBedrms']]))


    ## 8️⃣ K-Nearest Neighbors (KNN) Regression - Using 'latitude' & 'longitude' with Manhattan Distance
    knn = KNeighborsRegressor(n_neighbors=10, metric='manhattan')
    knn.fit(train_set[['latitude', 'longitude']], train_set["MedHouseVal"])
    rmses[7, i] = root_mean_squared_error(y_test, knn.predict(test_set[['latitude', 'longitude']]))

# Display the RMSE results
print("RMSE values for each model across folds:\n", rmses)
print("\nMean RMSE for each model:\n", np.mean(rmses, axis=1))
```
